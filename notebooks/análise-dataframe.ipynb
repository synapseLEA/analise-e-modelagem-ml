```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diretorio_notebook = os.path.dirname(os.path.abspath('análise-dataframe.ipynb'))
diretorio_raiz = os.path.dirname(diretorio_notebook)
caminho_csv = os.path.join(
    diretorio_raiz,
    'banco-dados',
    'bruto',
    'diabetes_dataset.csv'
)
df = pd.read_csv(caminho_csv)
```

```python
colunas_numericas = df.select_dtypes(exclude='object').columns
colunas_categoricas = df.select_dtypes(include='object').columns
```

```python
fig, axs = plt.subplots(
    nrows=len(colunas_categoricas),
    ncols=3,
    figsize=(15, 7 * len(colunas_categoricas))
)
for index, coluna in enumerate(colunas_categoricas):
    grafico = sns.countplot(
        x=df[coluna],
        ax=axs[index][0]
    )
    for valor in grafico.containers:
        axs[index][0].bar_label(valor)
    axs[index][0].set_title(
        f'Distribuição dos dados da coluna {coluna}'
    )
    axs[index][0].set_ylabel(
        "Frequência dos casos"
    )
    axs[index][0].set_xlabel(
        f"Coluna {coluna}"
    )
    nomes_variaveis = df[coluna].value_counts().index.to_list()
    print(nomes_variaveis)
    valores = df[coluna].value_counts().values.tolist()
    print(valores)
    maior_valor = valores[0]
    explode_list=[0.1 if valor == maior_valor else 0.0 for valor in valores]
    print(explode_list)
    axs[index][1].pie(
        x = valores,
        labels=nomes_variaveis,
        explode=explode_list,
        autocpt = '%1.1f%%'
    )
#plt.show()
```
